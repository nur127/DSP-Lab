```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Settings
# -----------------------------
f_signal = 20        # original signal frequency (Hz)
fs = 25              # sampling frequency (CHANGE THIS VALUE)
duration = 1         # 1 second duration

# -----------------------------
# 1. Continuous Signal
# -----------------------------
t = np.linspace(0, duration, 5000)               # dense timeline
x = np.sin(2 * np.pi * f_signal * t)            # continuous signal

# -----------------------------
# 2. Sampled Signal
# -----------------------------
Ts = 1 / fs
n = np.arange(0, duration, Ts)                   # sample times
x_sampled = np.sin(2 * np.pi * f_signal * n)    # samples

# -----------------------------
# 3. Reconstructed (Aliased) signal
# -----------------------------
t_interp = np.linspace(0, duration, 2000)
x_alias = np.interp(t_interp, n, x_sampled)

# -----------------------------
# Plotting (3 Subplots)
# -----------------------------
plt.figure(figsize=(12, 9))

# --- Subplot 1: Original ---
plt.subplot(3, 1, 1)
plt.plot(t, x)
plt.title(f"Original Continuous Signal ({f_signal} Hz)")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)

# --- Subplot 2: Sampled Points ---
plt.subplot(3, 1, 2)
plt.stem(n, x_sampled, linefmt='r-', markerfmt='ro', basefmt='k')
plt.title(f"Sampled Signal at fs = {fs} Hz")
plt.xlabel("Time (s)")
plt.ylabel("Samples")
plt.grid(True)

# --- Subplot 3: Aliased Signal ---
np.interp
plt.plot(t_interp, x_alias)
plt.title("Reconstructed (Aliased) Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)

plt.tight_layout()
plt.show()
```
