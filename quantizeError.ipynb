```python
import numpy as np
import matplotlib.pyplot as plt

# -------------------------------
# 1. Create Original Continuous Signal
# -------------------------------
t = np.linspace(0, 1, 1000)        # continuous time axis
x = 0.8 * np.sin(2*np.pi*5*t)      # original signal (amplitude < 1 so no clipping)

# -------------------------------
# 2. Quantization Settings
# -------------------------------
bits = 3                           # number of bits
levels = 2**bits                   # total quantization levels
xmin, xmax = -1, 1                 # quantizer range
delta = (xmax - xmin) / levels     # step size (quantization interval)

# -------------------------------
# 3. Perform Quantization
# -------------------------------
# Map x to nearest quantization interval
q = np.round((x - xmin) / delta) * delta + xmin

# Ensure quantized values remain inside the range
q = np.clip(q, xmin, xmax - delta)

# -------------------------------
# 4. Quantization Error
# -------------------------------
error = x - q

# -------------------------------
# 5. Plotting Using Subplots
# -------------------------------
plt.figure(figsize=(12,8))

# ---- Subplot 1: Original Signal ----
plt.subplot(3,1,1)
plt.plot(t, x, label='Original Signal', color='blue')
plt.title("Original Continuous Signal")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid(True)

# ---- Subplot 2: Quantized Signal ----
plt.subplot(3,1,2)
plt.step(t, q, where='mid', label='Quantized Signal', color='red')
plt.title("Quantized Signal")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid(True)

# ---- Subplot 3: Quantization Error ----
plt.subplot(3,1,3)
plt.plot(t, error, label='Quantization Error', color='green')
plt.title("Quantization Error (x - q)")
plt.xlabel("Time")
plt.ylabel("Error")
plt.grid(True)

plt.tight_layout()
plt.show()
```
