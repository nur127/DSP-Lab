```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
x= [1,1,2,2]
h=[1,2,3,4]
N = len(x)
M = len(h)

out_putlength = N+M-1
y = [0]*out_putlength;
print(y);
```

```python
for n in range(out_putlength):
    sum_val = 0 ;
    for k in range(N):
        if 0<= n-k < M : 
            sum_val += x[k] * h[n-k]
    y[n] = sum_val ;
    
print(f"Convulution : ", y);
```
