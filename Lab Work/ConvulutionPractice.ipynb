```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = 100
t = np.linspace(0,1,1000)
x_t = np.sin(2*np.pi*f*t)
fs = 250
Ts = 1/fs
p = np.arange(0,1,Ts)
x = np.sin(2*np.pi*f*p)
delay = 10
h = np.concatenate((np.zeros(delay), x[:-delay]))
```

```python
N = len(x)
M = len(h)
outputLength = N+M-1;
y = [0]*outputLength;
print(y)
```

```python
for n in range(outputLength):
    sum_value = 0 
    for k in range(N):
        if 0<=n-k<M:
            sum_value+=x[k] * h[n-k]
    y[n] = sum_value
    
print(y)
```

```python
plt.figure(figsize=(20,12))
plt.subplot(4,1,1)
plt.plot(t,x_t)
plt.subplot(4,1,2)
plt.stem(p,x,basefmt='k-',linefmt='r-',markerfmt='ro-')
plt.subplot(4,1,3)
plt.stem(p,h,basefmt='k-',linefmt='g-',markerfmt='go-')
plt.subplot(4,1,4)
plt.stem(np.arange(len(y)),y,basefmt='k-',linefmt='m-',markerfmt='mo-')
```
