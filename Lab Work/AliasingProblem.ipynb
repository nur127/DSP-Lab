```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0,1,1000)
x_t = np.sin(2*np.pi*5*t);
```

```python
#Bad Sampling 
fs1 = 8 
Ts1 = 1/fs1
n1 = np.arange(0,1,Ts1)
x_n1 =np.sin(2*np.pi*5*n1)
```

```python
#Good Sampling 
fs2 = 15 
Ts2 = 1/fs2
n2 = np.arange(0,1,Ts2)
x_n2 =np.sin(2*np.pi*5*n2)
```

```python
plt.figure()
plt.plot(t,x_t,label="Continuous Time Signal")
plt.stem(n1,x_n1,markerfmt='ro' ,basefmt='k-',linefmt='r-',label="Bad Sampling")
plt.stem(n2,x_n2,linefmt='g-',markerfmt='go',basefmt='k-',label="Good Sampling")
plt.show()
```
