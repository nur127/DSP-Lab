```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
n = np.arange(-11,10,1)
```

```python
u = np.where(n>= -5 ,1,0)
```

```python
d = np.where(n== 0 ,1,0)
```

```python
r = n*np.where(n>= 0 ,1,0)
```

```python
plt.figure (figsize=(40,20))
plt.subplot(3,1,1)
plt.stem(n,u,linefmt='r-',markerfmt='ro',basefmt='k-', label="Unit signal")
plt.subplot(3,1,2)
plt.stem(n,d,linefmt='r-',markerfmt='ro',basefmt='k-', label="Unit signal")
plt.subplot(3,1,3)
plt.stem(n,r,linefmt='r-',markerfmt='ro',basefmt='k-', label="Unit signal")
```
