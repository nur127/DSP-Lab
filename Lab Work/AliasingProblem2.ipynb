```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
f = 20 ;
fs = 200;
```

```python
t = np.linspace(0,1,5000)
x = np.cos(2*np.pi*f*t)
```

```python
#Sample signal 
Ts = 1/fs;
n = np.arange(0,1,Ts)
x_n = np.cos(2*np.pi*f*n)
```

```python
t_interp = np.linspace(0,1,5000)
x_alias = np.interp(t_interp,n,x_n)
```

```python
plt.figure(figsize=(60,40))
plt.subplot(3,1,1)
plt.plot(t,x)
plt.title(f"Continous Signal of {f} frequency")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid(True)
plt.subplot(3,1,2)
plt.stem(n,x_n,linefmt='r-',markerfmt='ro',basefmt='k-')
plt.title(f"Discrete Signal of {fs} frequency")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid(True)
plt.subplot(3,1,3)
plt.plot(t_interp,x_alias)
plt.title(f"Discrete Signal of {fs} frequency")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid(True)
```

```python
f = 100
t = np.linspace(0,1,5000)
x_t = np.sin(2 * np.pi*f *t)

fs = 150
Ts = 1/fs
n = np.arange(0,1,Ts)
x_n = np.sin(2 * np.pi*f *n)

t_r = np.linspace(0,1,5000)
x_r = np.interp(t_r,n,x_n)

plt.figure(figsize=(20,12))
plt.subplot(3,1,1)
plt.plot(t,x_t)
plt.subplot(3,1,2)
plt.stem(n,x_n,basefmt='k-',markerfmt='ro',linefmt='r-')
plt.subplot(3,1,3)
plt.plot(t_r,x_r)
```
