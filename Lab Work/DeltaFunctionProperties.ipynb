```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
n = np.arange(-11,10,1)
a = 2 
```

```python
d = np.where(n==0,1,0)
```

```python
scaleSignal = (1/abs(a) ) * np.where(n==0,1,0)
```

```python
plt.figure(figsize=(30,20))
plt.subplot(2,1,1)
plt.stem(n,d,linefmt='r-',basefmt='k-',markerfmt='ro')
plt.subplot(2,1,2)
plt.stem(n,scaleSignal,linefmt='g-',basefmt='k-',markerfmt='go')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# time index
n = np.arange(-10, 11, 1)

# original delta signal δ(n)
delta_n = np.where(n == 0, 1, 0)

# scaling factor
a = 2

# δ(a n) → still impulse at n=0, but amplitude changes
delta_an = (1/abs(a)) * np.where(n == 0, 1, 0)

# plot
plt.figure(figsize=(8,5))

plt.stem(n, delta_n, linefmt='b-', markerfmt='bo', basefmt='k-', label="δ(n)")
plt.stem(n, delta_an, linefmt='r-', markerfmt='ro', basefmt='k-', label="δ(an) = 1/|a| δ(n)")

plt.title("Scaling Property of Delta Signal")
plt.xlabel("n")
plt.ylabel("Amplitude")
plt.legend()
plt.grid()
plt.show()
```
