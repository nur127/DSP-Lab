```python
# x = np.array([1,2,3,4])
x = np.array([5,0,-5,0])
N = len(x)
```

```python
X = np.zeros(N,dtype=complex)
```

```python
for k in range(N):
    for n in range(N):
        X[k] += x[n] * np.exp(-1j * 2*np.pi*n*k/N)
```

```python
magnitude = np.abs(X)
phase = np.angle(X)
print(magnitude , phase)
```

```python
plt.figure(figsize=(20,4))
plt.subplot(2,1,1)
plt.stem(magnitude)
plt.subplot(2,1,2)
plt.stem(phase)
```

```python
# IDFT 
X_r = np.zeros(N,dtype=complex)
```

```python
for n in range(N):
    for k in range(N):
        X_r[n] += X[k] * np.exp(1j * 2*np.pi*n*k/N)
    X_r[n] = (X_r[n]/N )
    
print(X_r)
```
