```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# For Sin Wave 
f = 20 
fs = 60
Ts = 1/fs;
t  = np.linspace(0,1,1000)
x_t = np.sin(2 * np.pi *f*t)
p = np.arange(0,1,Ts)
x = np.sin(2 *np.pi * f * p)
delay = 10 ;
# h = np.roll(x,delay)
h = np.concatenate((np.zeros(delay), x[:-delay]))
y=np.correlate(x,h,'full')
```

```python
# x = [1,1,2,2]
# h = [1,2,3,4]
```

```python
N = len(x)
M = len(h)
print(N,M)
# Length of O/p
output_size = N+M-1;
r = [0]*output_size
# print(r)
```

```python
for n in range(-(M-1),N):
    print(n)
    sum_val = 0 ;
    for k in range(N):
        if 0 <= k-n <M:
            sum_val += h[k-n]*x[k]
    r[n+M-1] = sum_val
    
# r = r[::-1]
print(r)

    
```
