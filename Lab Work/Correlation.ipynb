```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# For Sin Wave 
f = 20 
fs = 60
Ts = 1/fs;
t  = np.linspace(0,1,1000)
x_t = np.sin(2 * np.pi *f*t)
p = np.arange(0,1,Ts)
x = np.sin(2 *np.pi * f * p)
delay = 10 ;
# h = np.roll(x,delay)
h=np.sin(2 *np.pi * f * (p-delay*Ts))
y=np.correlate(x,h,'full')
```

```python
# x = [1,1,2,2]
# h = [1,2,3,4]
```

```python
N = len(x)
M = len(h)
print(N,M)
# Length of O/p
output_size = N+M-1;
r = [0]*output_size
# print(r)
```

```python
for n in range(-(N-1),M):
    sum_val = 0 ;
    for k in range(N):
        if 0 <= k-n <M:
            sum_val += h[k-n]*x[k]
    r[n+N-1] = sum_val
    
# r = r[::-1]
print(r)
    
```

```python
plt.figure(figsize=(20,12))
plt.subplot(4,1,1)
plt.plot(t,x_t)
plt.title("Continous signal ")
plt.subplot(4,1,2)
plt.stem(p,x,linefmt='g-', markerfmt='go',basefmt='k-' )
plt.title("discrete signal")
plt.subplot(4,1,3)
plt.stem(p,h,linefmt='r-', markerfmt='ro',basefmt='k-' )
plt.title("Impulse signal")
plt.subplot(4,1,4)
plt.stem(np.arange(-M+1,N),y,linefmt='b-', markerfmt='bo',basefmt='k-' )
plt.title("Cross-correlation")
plt.xlabel("Time")
plt.ylabel("Amplitude")
```
