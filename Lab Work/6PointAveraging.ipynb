```python
import numpy as np
import matplotlib.pyplot as plt

# Input signal
x = np.array([1, 2, 3, 4, 5, 6, 7, 8])

# 6-point averaging filter
h = np.ones(6) / 6
print(h)

Nx = len(x)
Nh = len(h)

y = np.zeros(Nx + Nh - 1)

# Manual convolution
for n in range(len(y)):
    for k in range(Nx):
        if 0 <= n-k < Nh:
            y[n] += x[k] * h[n-k]

print("Output:", y)

# Plot
plt.stem(y)
plt.title("6-point Averaging Filter Output")
plt.show()
```
