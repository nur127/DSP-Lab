```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = 100
t = np.linspace(0,1,1000)
x_t = np.cos(2 * np.pi*f *t)

fs = 200
Ts = 1/fs
n = np.arange(0,1,Ts)
x_n = np.cos(2 * np.pi*f *n)

t_r = np.linspace(0,1,1000)
x_r = np.interp(t_r,n,x_n)

plt.figure(figsize=(20,12))
plt.subplot(3,1,1)
plt.plot(t,x_t)
plt.subplot(3,1,2)
plt.stem(n,x_n,basefmt='k-',markerfmt='ro',linefmt='r-')
plt.subplot(3,1,3)
plt.plot(t_r,x_r)
```
