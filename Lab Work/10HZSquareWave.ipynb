```python
import numpy as np
import matplotlib.pyplot as plt

# =========================
# PARAMETERS
# =========================
f = 10          # frequency = 10 Hz
fs = 100        # sampling frequency
t = np.arange(0, 1, 1/fs)

# =========================
# GENERATE SINE WAVE
# =========================
x = np.sin(2 * np.pi * f * t)

# =========================
# CONVERT TO SQUARE WAVE
# =========================
square = []

for val in x:
    if val >= 0:
        square.append(1)
    else:
        square.append(-1)

square = np.array(square)

# =========================
# PLOT
# =========================
plt.figure(figsize=(10,4))
plt.step(t, square)
plt.title("10 Hz Square Wave")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.ylim(-1.5, 1.5)
plt.grid()
plt.show()
```
