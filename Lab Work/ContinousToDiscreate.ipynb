```python
import numpy as np 
import matplotlib.pyplot as plt  
```

```python
t = np.linspace(0,1,1000)
x_t = np.sin(2*np.pi*5*t)
```

```python
fs = 20
Ts = 1/fs 
```

```python
n = np.arange(0,1,Ts)
x_n = np.sin(2*np.pi*5*n)
```

```python
plt.figure()
plt.plot (t,x_t,label="Continuous Time Signal")
plt.stem(n,x_n,linefmt='g-',markerfmt='ro',basefmt='k-',label='Discrete Signal')
plt.title("Continuous to Discrete Signal")
plt.xlabel("Time")
plt.ylabel("Amplitude")
# plt.legend()
plt.show()
```
