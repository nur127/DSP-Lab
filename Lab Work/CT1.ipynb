```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3])
h = np.array([1,1,1])
```

```python
N = len(x)
M = len(h)
OutputLength = N+M-1
y = [0]*OutputLength
```

```python
for n in range(OutputLength):
    sum_value = 0
    for k in range(N):
        if 0<=n-k<M :
            sum_value+=x[k]*h[n-k]
    y[n]=sum_value


print(f"Convulution Result in Time Domain : {y}")
    
```

```python
def DFT(x):
    N = len(x)
    X = np.zeros(N,dtype=complex)
    for k in range(N):
        for n in range(N):
            X[k] += x[n] * np.exp(-1j*2*np.pi*n*k/N)
    return X;

    
```

```python
def IDFT(x):
    N = len(x)
    Xi = np.zeros(N,dtype=complex)
    for n in range(N):
        for k in range(N):
            Xi[n] += x[k] * np.exp(1j*2*np.pi*n*k/N)
        Xi[n] = Xi[n]/N
    return Xi;
```

```python
x_pad = np.concatenate((x,np.zeros(OutputLength-N)))
h_pad = np.concatenate((h,np.zeros(OutputLength-M)))
```

```python
X = DFT(x_pad)
H = DFT(h_pad)
Y = X*H
print(Y)
```

```python
Xidft = IDFT(Y)
print(np.abs(Xidft))
```
