```python
import numpy as np

# Example coefficients
b = [1, 0.5]     # input coefficients
a = [1, -0.3]    # output coefficients (a[0] always 1)

x = np.array([1,2,3,4,5])
N = len(x)

y = np.zeros(N)

for n in range(N):
    # input part
    for k in range(len(b)):
        if n-k >= 0:
            y[n] += b[k] * x[n-k]
    
    # output part
    for k in range(1, len(a)):
        if n-k >= 0:
            y[n] -= a[k] * y[n-k]

print("Output:", y)
```
