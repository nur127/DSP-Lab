```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# Sin 
f = 30
t = np.linspace(0,1,1000)
x_t = np.sin(2 * np.pi * f *t)
fs = 100
Ts = 1/fs
n = np.arange(0,1,Ts)
x = np.sin(2 * np.pi * f * n)
N=100
```

```python
# x = np.array([1,2,3,4])
# x = np.array([5,0,-5,0])
# N = len(x)
```

```python
X = np.zeros(N,dtype=complex)
print(X)
```

```python
for k in range(N):
    for n in range(N):
        X[k] += x[n] *np.exp(-1j*2*np.pi*n*k / N)
```

```python
print("DFT Result : ")
print(X)
```

```python
magnitude = np.abs(X)
phase = np.angle(X)
```

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.stem(magnitude)
plt.title("Magnitude Spectrum")
plt.xlabel("k")
plt.ylabel("|X(k)|")

plt.subplot(1,2,2)
plt.stem(phase)
plt.title("Phase Spectrum")
plt.xlabel("k")
plt.ylabel("Angle")

plt.tight_layout()
plt.show()
```

```python
x_idft = np.zeros(N,dtype=complex)

for n in range(N):
    for k in range(N):
        x_idft [n] += X[k] * np.exp(1j * 2 * np.pi * k * n/N)
    x_idft[n] = x_idft[n]/N
    

print("Reconstructed Signal:")
print(x_idft.real)
```

```python
#Rectenguilar 
x = np.array([1, 2, 3, 4])

# Rectangular window (all 1)
w_rect = np.ones(len(x))

x_rect = x * w_rect

print("Rectangular Window Output:", x_rect)
N = len(x)

w_hamming = 0.54 - 0.46 * np.cos(2 * np.pi * np.arange(N) / (N - 1))

x_hamming = x * w_hamming

print("Hamming Window Output:", x_hamming)
plt.stem(w_hamming)
plt.title("Hamming Window")
plt.show()
```
