```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fs = 8000
fc = 1000
N = 21
M = (N-1)//2
```

```python
n = np.arange(N)
print(n)
```

```python
# Impulse Response 
h_d = 2 * fc/fs *np.sinc(2*(fc/fs)*(n-M))
```

```python
# Hamming window
w = .56 -.46*np.cos(2*np.pi*n/(N-1))
```

```python
# Original Impulse 
h = h_d * w 
```

```python
H = np.zeros(N, dtype=complex)
```

```python
for k in range (N) : 
    for i in range (N):
        H[k] += h[i] * np.exp(-1j *2*np.pi*i*k/N)
```

```python
magnitude = np.abs(H)
phase = np.angle(H)
```

```python
plt.figure(figsize=(25,12))
plt.subplot(5,1,1)
plt.stem(n,h_d)
plt.subplot(5,1,2)
plt.stem(n,h)
plt.subplot(5,1,3)
plt.stem(n,H)
plt.subplot(5,1,4)
plt.stem(magnitude)
plt.subplot(5,1,5)
plt.stem(phase)

```
